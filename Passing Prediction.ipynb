```python
import numpy as np

study_hours = np.array([1, 2, 3, 4, 5])
exam_scores = np.array([1, 0, 0, 1, 1])

#initialize parameters
w = 0.0  # slope    
b = 0.0  # intercept

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def predict_probability(study_hours, w, b):
    x = w * study_hours + b
    return sigmoid(x)

def lose_function(end_result, model_result):
    return -np.mean(end_result * np.log(model_result + 1e-9) + (1 - end_result) * np.log(1 - model_result + 1e-9))

learning_rate = 0.1
echoes = 500

for i in range(echoes):
    model_result = predict_probability(study_hours, w, b)

    dw =np.mean((model_result - exam_scores) * study_hours)
    db =np.mean((model_result- exam_scores))

    w -= learning_rate * dw
    b -= learning_rate * db 
   
    if i % 50 == 0:
        loss = lose_function(exam_scores, model_result)
        print("ECHO", i , "LOSS", loss)


def predict(x): 
        prob = predict_probability(x, w, b)
        return (prob >= 0.5).astype(int)
        
study_hours = 1.1

prob = predict_probability(study_hours, w, b)
result = predict(study_hours)
print("probability of passing" , prob)
print("prediction (1 = pass, 0 = fail):", result)
    
    
    
```
